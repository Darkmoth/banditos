```python
import pandas as pd
import numpy as np
import datetime
import calendar
```

```python
currTime = datetime.datetime.now()
currDate = currTime.date()
```

```python
daynames = [calendar.day_name[x] for x in range(7)]
weekdays = pd.DataFrame({'Weekday': daynames})
weekdays['Date'] = currDate
```

```python
weekdays
```

```python
goals = pd.DataFrame({'Goal': range(1,22)})
goals['Date'] = currDate
```

```python
goals
```

```python
goal_schedule = weekdays.merge(goals, how='left', on = 'Date')
goal_schedule['Group'] = (goal_schedule['Goal'] - 1) // 7
goal_schedule['Num Completed'] = np.nan
#days_and_stores.drop('key',1, inplace=True)
```
